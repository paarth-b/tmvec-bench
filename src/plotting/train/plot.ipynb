# Foundation model training

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
plt.rcParams['svg.fonttype'] = 'none'
```

```python
methods = ['Lobster-24M', 'Lobster-150M', 'ProtT5-XL', 'Ankh-base', 'Ankh-large', 'ProtMamba'] 
```

```python
hours = [0.11532, 0.19753, 1.55679, 1.70257, 1.73112, 6.53164]
```

Results were taken from:

- https://api.wandb.ai/links/tmvec-retrain/vzlxxegf

```python
cmap = plt.get_cmap('tab10')
```

```python
plt.figure(figsize=(6, 3))
xran = range(len(methods))
plt.bar(x=methods, height=hours, color=[cmap(i) for i in xran])
plt.xticks(xran, methods, rotation=45, ha='right')
plt.grid(True)
plt.gca().set_axisbelow(True)
plt.xlabel('Foundation model')
plt.ylabel('Training time (hr)')
plt.savefig('plots/training.svg')
```
