Results

```python
df = pd.read_table('results.tsv')
df
```

```python
methods = df.columns[2:].tolist()
print(methods)
```

```python
lengths = df['Sequence Length'].unique()
lengths
```

```python
batches = df['Batch Size'].unique()
batches
```

```python
df_melt = pd.melt(df, ['Sequence Length', 'Batch Size'], var_name='Method', value_name='Ms per protein')
df_melt
```

Line plots

```python
fig, axes = plt.subplots(4, 2, figsize=(12, 12), sharex=True)
for i, length in enumerate(lengths):
    ax = axes[i // 2, i % 2]
    data = df_melt.query(f'`Sequence Length` == {length}')
    for method in methods:
        ax.plot('Batch Size', 'Ms per protein', data=data.query(f'Method == "{method}"'), marker='o', label=method)
    if i == 0:
        ax.legend()
    if i % 2 == 0:
        ax.set_ylabel('Ms per protein')
    if i // 2 == 3:
        ax.set_xlabel('Batch size')
    ax.set_title(f'{length} aa')
fig.tight_layout()
fig.savefig('plots/infer_line.svg')
```

Bar plots

```python
fig, axes = plt.subplots(4, 2, figsize=(12, 12), sharex=True)
for i, length in enumerate(lengths):
    ax = axes[i // 2, i % 2]
    data = df_melt.query(f'`Sequence Length` == {length}')
    sns.barplot(data=data, x='Batch Size', y='Ms per protein', hue='Method', legend=not i, ax=ax)
    if i % 2:
        ax.set_ylabel(None)
    ax.set_title(f'{length} aa')
fig.tight_layout()
fig.savefig('plots/infer_bar.svg')
```
